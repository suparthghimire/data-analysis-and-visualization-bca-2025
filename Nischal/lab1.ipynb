```python
import matplotlib.pyplot as plt

x = list(range(1, 11))
y = [i**2 for i in x]

plt.plot(x, y, marker='o')
plt.title("Square of Numbers from 1 to 10")
plt.xlabel("Number")
plt.ylabel("Square")
plt.grid(True)
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x, sin_y, label='sin(x)', linestyle='--', color='blue')
plt.plot(x, cos_y, label='cos(x)', linestyle='-', color='green')
plt.title("Sin and Cos Functions")
plt.xlabel("x (radians)")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

colors = ['red' if val < 0 else 'blue' for val in x]

plt.scatter(x, y, c=colors, alpha=0.6)
plt.title("Random Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

```python
provinces = ['Province 1', 'Madhesh', 'Bagmati', 'Gandaki', 'Lumbini']
populations = [5400000, 6200000, 5800000, 2500000, 4900000]
colors = ['skyblue', 'orange', 'green', 'red', 'purple']

plt.bar(provinces, populations, color=colors)
plt.title("Population by Province")
plt.xlabel("Province")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.show()
```

```python
explode = [0, 0.1, 0, 0, 0]  # Explode Madhesh
plt.pie(populations, labels=provinces, autopct='%1.1f%%', explode=explode, colors=colors)
plt.title("Population Distribution by Province")
plt.show()
```

```python
import pandas as pd

dates = pd.date_range(start="2023-01-01", periods=30)
temps = np.random.normal(loc=20, scale=5, size=30)

df_temp = pd.DataFrame({'Date': dates, 'Temperature': temps})
df_temp.set_index('Date', inplace=True)

df_temp.plot(figsize=(10, 4), title="Daily Temperature for 30 Days")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()
```

```python
import mplfinance as mpf

dates = pd.date_range(start="2023-01-01", periods=30)
ohlc_data = pd.DataFrame({
    'Open': np.random.rand(30) * 100 + 100,
    'High': np.random.rand(30) * 100 + 150,
    'Low': np.random.rand(30) * 100 + 80,
    'Close': np.random.rand(30) * 100 + 100,
    'Volume': np.random.randint(100000, 200000, size=30)
}, index=dates)

mpf.plot(ohlc_data, type='candle', style='charles',
         title='Sample Candlestick Chart', figsize=(10, 6), volume=True)
```

```python
from ipywidgets import interact, widgets

def plot_functions(show_sin, show_cos, show_tan, color, points):
    x = np.linspace(0, 2 * np.pi, points)
    plt.figure(figsize=(10, 5))

    if show_sin:
        plt.plot(x, np.sin(x), label='sin(x)', color=color)
    if show_cos:
        plt.plot(x, np.cos(x), label='cos(x)', linestyle='--', color=color)
    if show_tan:
        y = np.tan(x)
        y[np.abs(y) > 10] = np.nan  # Avoid vertical lines
        plt.plot(x, y, label='tan(x)', linestyle=':', color=color)

    plt.legend()
    plt.title("Trigonometric Functions")
    plt.xlabel("x (radians)")
    plt.ylabel("Function value")
    plt.grid(True)
    plt.show()

interact(plot_functions,
         show_sin=widgets.Checkbox(value=True, description='sin(x)'),
         show_cos=widgets.Checkbox(value=False, description='cos(x)'),
         show_tan=widgets.Checkbox(value=False, description='tan(x)'),
         color=widgets.ColorPicker(value='blue', description='Color'),
         points=widgets.IntSlider(value=100, min=50, max=300, step=50, description='Points'))
```
