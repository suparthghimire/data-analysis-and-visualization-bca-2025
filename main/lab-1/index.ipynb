# Visualization

```python
plt.figure(figsize=(8, 6))

plt.plot(housing_test_data_frame["population"], housing_test_data_frame["total_rooms"], label="Total Rooms", color="blue")
plt.plot(housing_test_data_frame["population"], housing_test_data_frame["households"], label="Households", color="green")

plt.xlabel("Population")
plt.ylabel("Values")
plt.title("Population vs Total Rooms & Households")
plt.legend()
plt.grid(True)
plt.show()
```

```python
housing_test_data_frame.columns
```

```python
import matplotlib.pyplot as plt

# Top 5 entries
sizes = housing_test_data_frame["total_rooms"][:5]
labels = [f"Entry {i}" for i in sizes.index]  # e.g., Entry 0, Entry 1, etc.

plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(
    sizes,
    labels=None,
    autopct='%1.1f%%',
    startangle=140
)

plt.legend(wedges, labels, title="Data Entries", loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("Total Rooms for Top 5 Entries")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Total rooms by Income Range

```python
# Create income bins
housing_test_data_frame["income_group"] = pd.cut(
    housing_test_data_frame["median_income"],
    bins=[0, 2, 4, 6, 8, 15],
    labels=["Very Low", "Low", "Medium", "High", "Very High"]
)

# Group and sum
grouped = housing_test_data_frame.groupby("income_group")["total_rooms"].sum()
print(grouped)

# Pie chart
plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(
    grouped.values,
    labels=None,
    autopct='%1.1f%%',
    startangle=140
)
plt.legend(wedges, grouped.index, title="Income Group", loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("Total Rooms by Income Group")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
