```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head()
```

1. Histograms (for Numerical Data)

Standard Histogram – Age

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=False)
plt.title("Standard Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

1.2 Normalized Histogram – Fare

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=30, stat='density')
plt.title("Normalized Histogram of Fare")
plt.xlabel("Fare")
plt.ylabel("Density")
plt.show()
```

1.3 Cumulative Histogram – Age

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, cumulative=True)
plt.title("Cumulative Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Cumulative Count")
plt.show()
```

1.4 KDE / Density Plot – Fare

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['Fare'], fill=True)
plt.title("Density Plot of Fare")
plt.xlabel("Fare")
plt.ylabel("Density")
plt.show()
```

**2. Bar Plots (for Categorical Data)**

2.1 Vertical Bar Chart – Sex

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Sex', data=df)
plt.title("Passenger Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

   2.2 Stacked Bar Chart – Sex vs. Survival

```python
pd.crosstab(df['Sex'], df['Survived']).plot(kind='bar', stacked=True)
plt.title("Survival by Sex (Stacked)")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(["Did not survive", "Survived"])
plt.show()
```

2.3 Grouped Bar Chart – Pclass vs. Survived

```python
pd.crosstab(df['Pclass'], df['Survived']).plot(kind='bar', stacked=False)
plt.title("Survival by Passenger Class (Grouped)")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(["Did not survive", "Survived"])
plt.show()
```

2.4 Lollipop Chart – Embarked

```python
counts = df['Embarked'].value_counts()
plt.figure(figsize=(6,4))
plt.stem(counts.index, counts.values, basefmt=" ")
plt.title("Lollipop Chart of Embarked Locations")
plt.xlabel("Port")
plt.ylabel("Count")
plt.show()
```

3. Pie Charts (for Category Proportion)

3.1 Standard Pie Chart – Sex

```python
df['Sex'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel('')
plt.show()
```

3.2 Donut Chart – Embarked

```python
embarked_counts = df['Embarked'].value_counts()
plt.pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%', wedgeprops=dict(width=0.4))
plt.title("Donut Chart of Embarked")
plt.show()
```

3.3 Exploded Pie Chart – Pclass

```python
pclass_counts = df['Pclass'].value_counts()
explode = (0.1, 0, 0)
plt.pie(pclass_counts, labels=pclass_counts.index, explode=explode, autopct='%1.1f%%')
plt.title("Exploded Pie Chart of Passenger Classes")
plt.show()
```

3.4 Nested Pie (Sunburst style) – Sex and Survived

```python
import plotly.express as px

fig = px.sunburst(df, path=['Sex', 'Survived'], values=None)
fig.update_layout(title="Nested Pie (Sunburst) of Sex and Survival")
fig.show()
```

4. Box Plots (for Statistical Summary)

4.1 Standard Box Plot – Age

```python
plt.figure(figsize=(6,5))
sns.boxplot(y='Age', data=df)
plt.title("Box Plot of Age")
plt.show()
```

4.2 Notched Box Plot – Fare

```python
plt.figure(figsize=(6,5))
sns.boxplot(y='Fare', data=df, notch=True)
plt.title("Notched Box Plot of Fare")
plt.show()
```

4.3 Violin Plot – Age

```python
plt.figure(figsize=(6,5))
sns.violinplot(y='Age', data=df)
plt.title("Violin Plot of Age")
plt.show()
```

4.4 Side-by-side Box Plot – Age by Sex

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='Sex', y='Age', data=df)
plt.title("Box Plot of Age by Sex")
plt.show()
```
