```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

# Load iris dataset
iris = load_iris(as_frame=True)
df_iris = iris.frame
df_iris['species'] = df_iris['target'].map(dict(enumerate(iris.target_names)))

# Bar chart
df_iris['species'].value_counts().plot(kind='bar', color='skyblue', title='Iris Species Count')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()


```

```python
# Pie chart
df_iris['species'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Species Distribution')
plt.ylabel('')
plt.show()
```

2. Ordinal Data

```python
# Ordinal Data - Simulated Ratings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
df_rating = pd.DataFrame({
    'rating': np.random.choice([1, 2, 3, 4, 5], size=500, p=[0.1, 0.2, 0.3, 0.25, 0.15])
})

# Ordered Bar Chart
plt.figure()
df_rating['rating'].value_counts().sort_index().plot(kind='bar', color='orange')
plt.title('Customer Ratings')
plt.xlabel('Rating (1 = Worst, 5 = Best)')
plt.ylabel('Count')
plt.show()

# Strip Plot
plt.figure()
sns.stripplot(x='rating', data=df_rating, order=[1,2,3,4,5], jitter=True)
plt.title('Customer Rating Distribution')
plt.show()
```

3. Interval Data
Use data with equal intervals but no true zero (e.g., temperature in C).
Show at least 3 visualizations:
Line/bar chart
Heatmap or gradient color chart
Emphasize relative difference

```python

# Simulated temperature dataset
dates = pd.date_range(start="2024-01-01", periods=30)
temperatures = np.random.normal(loc=20, scale=5, size=30)
df_temp = pd.DataFrame({'Date': dates, 'Temperature_C': temperatures})

# Line chart
plt.plot(df_temp['Date'], df_temp['Temperature_C'], color='green')
plt.title('Daily Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar chart
df_temp.set_index('Date')['Temperature_C'].plot(kind='bar', color='skyblue', figsize=(12,4))
plt.title('Temperature Variation')
plt.ylabel('C')
plt.show()

# Heatmap
df_temp['Day'] = df_temp['Date'].dt.day
sns.heatmap(df_temp.pivot_table(values='Temperature_C', index='Day'), cmap='coolwarm', annot=True)
plt.title('Temperature Heatmap')
plt.show()
```

4. Ratio Data

```python
# Simulated GDP data
countries = ['USA', 'China', 'India', 'Germany', 'Brazil']
gdp = [21.4, 14.3, 2.9, 3.8, 2.1]
df_gdp = pd.DataFrame({'Country': countries, 'GDP (Trillion USD)': gdp})

# Diverging bar chart
df_gdp_sorted = df_gdp.sort_values('GDP (Trillion USD)')
plt.barh(df_gdp_sorted['Country'], df_gdp_sorted['GDP (Trillion USD)'], color='steelblue')
plt.axvline(0, color='black')
plt.title('GDP of Selected Countries')
plt.xlabel('Trillion USD')
plt.show()




```

```python

population = [331, 1439, 1380, 83, 212]  # in millions
df_gdp['Population'] = population
plt.scatter(df_gdp['Population'], df_gdp['GDP (Trillion USD)'], s=df_gdp['GDP (Trillion USD)']*100, alpha=0.6)
plt.title('GDP vs Population')
plt.xlabel('Population (millions)')
plt.ylabel('GDP (Trillion USD)')
for i in range(len(df_gdp)):
    plt.text(df_gdp['Population'][i], df_gdp['GDP (Trillion USD)'][i], df_gdp['Country'][i])
plt.show()
```

```python
# Lollipop chart
plt.stem(df_gdp['Country'], df_gdp['GDP (Trillion USD)'], basefmt=" ")

plt.title('GDP Comparison (Lollipop Chart)')
plt.ylabel('GDP in Trillions')
plt.show()
```
